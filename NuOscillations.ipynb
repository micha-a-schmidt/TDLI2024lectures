```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['text.usetex']=True

color_cycle= plt.rcParams['axes.prop_cycle'].by_key()['color']

def P(L,theta,dms,E):
    return np.sin(theta)**2 * np.sin(1.27*dms*L/E)**2
```

```python

theta=70.0*np.pi/180.0
E=1.
dms=8e-5
maxL=(np.pi /2)* E /(1.27*dms)

x = np.linspace(0.0,3,100)
Pemu= [ P(L*maxL,theta,dms,E) for L in x]
Pee= [1-p for p in Pemu ]
```

```python
fig,ax = plt.subplots()
ax.plot(x,Pemu,label=r'P(\nu_e\to \nu_\mu)')
ax.plot(x,Pee,label=r'P(\nu_e\to \nu_e)',ls="--")
ax.set_ylabel(r'probability',fontsize=16)
ax.set_xlabel(r'$L/L_{\rm max}$',fontsize=16)
ax.set_xticks([0,0.5,1,1.5,2,2.5,3,3.5,4])
ax.set_yticks([0,0.25,0.5,0.75,1])
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16) 
ax.text(0.22,0.95,r'$P(\nu_e\to\nu_e)$',fontsize=16,color=color_cycle[1])
ax.text(0.22,0.05,r'$P(\nu_e\to\nu_\mu)$',fontsize=16,color=color_cycle[0])

ax.text(0.3,0.55,r'$L_{\rm max}$',fontsize=16)
ax.arrow(0.5,0.5,0.45,0,fc='black',ec='black',head_width=0.05,head_length=0.05)
ax.arrow(0.5,0.5,-0.45,0,fc='black',ec='black',head_width=0.05,head_length=0.05)

ax.text(2.85,0.55,r'$\sin^2\theta$',fontsize=16)
ax.arrow(3,0.63,0,0.32,fc='black',ec='black',head_width=0.05,head_length=0.05)
ax.arrow(3,0.5,0,-0.33,fc='black',ec='black',head_width=0.05,head_length=0.05)
plt.tight_layout()
plt.savefig("nuosc2.png") 
```
